## O que é estatística?
A estatística é a ciência que coleta, organiza, analisa e interpreta dados para auxiliar na tomada de decisões e previsão de eventos. Utiliza métodos matemáticos e probabilísticos para transformar dados brutos em informações úteis.
Na estatística temos variaveis, uma **variável** (em estatística) é qualquer característica que pode ser medida ou contada. Dividem-se em dois grandes grupos:
* **Variáveis Qualitativas (ou Categóricas):** Representam qualidades, rótulos ou categorias. Não são números com os quais fazemos contas matemáticas. Exemplo: cor dos olhos, cidade de residência, género.
* **Variáveis Quantitativas (ou Numéricas):** Representam quantidades e números reais. Exemplo: idade, salário, altura, número de filhos.

## Estatística Descritiva Básica
A estatística descritiva é o ramo da estatística que organiza, resume e apresenta dados brutos para torná-los compreensíveis, identificando padrões, tendências e variações. Ela não realiza inferências sobre populações maiores, focando apenas no conjunto de dados analisado através de medidas de resumo (média, mediana, moda), dispersão (desvio padrão, variância) e visualização gráfica.

#### O que é o Espaço Amostral?
Na estatística, o **espaço amostral** é simplesmente o conjunto de *todos os resultados possíveis* de uma variável ou de uma experiência. Se você jogar um dado normal de 6 faces, o espaço amostral (ou resultados possíveis) dele ao cair são: `[1, 2, 3, 4, 5, 6]`. Se jogar uma moeda o espaço amostral é `[Cara, Coroa]`.

Agora vamos calcular as medidas para as Horas_de_Sono dos alunos no nosso dataset.

```python
media = df['Horas_de_Sono'].mean() # calculando a média da coluna 'Horas_de_Sono'

mediana = df['Horas_de_Sono'].median() # calculando a mediana da coluna 'Horas_de_Sono'

moda = df['Horas_de_Sono'].mode()[0] # calculando a moda da coluna 'Horas_de_Sono'

print("Média:",media)
print("Mediana:",mediana)
print("Moda:",moda)
```
