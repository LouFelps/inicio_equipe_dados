## 🟢 Parte 6 — Condições (IF)

Agora Python começa a “pensar”.

**Aprende:**
- lógica
- indentação (conceito MUITO importante)

```python
idade = int(input("Idade: "))

if idade >= 18:
    print("Você é maior de idade")
else:
    print("Você é menor de idade")
```

### Mais condições

```python
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

### 🤖 Assistente simples

```python
def assistente():
    nome = input("Qual seu nome? ")
    idade = int(input("Sua idade: "))

    print("Olá,", nome)

    if idade >= 18:
        print("Você pode dirigir 🚗")
    else:
        print("Ainda não pode dirigir")

assistente()
```
