Machine Learning ou Aprendizado de Máquina é uma diferente forma de programação tradicional (onde o programador escreve um fluxo de regras de cima para baixo), no Machine Learning, algoritmos são expostos a grandes quantidades de dados e aprendem a identificar padrões e extrair informações para prever comportamentos e lidar com situações inesperadas (dados inéditos). Um exemplo prático dessa generalização são os chatbots, que conseguem formular respostas adequadas para perguntas que nunca viram antes.

<h3 style="color: #f686bd">Os Três Tipos Principais de Aprendizado de Máquina</h3>
A área de Machine Learning se divide basicamente em três categorias:
<div style="width: 380px; min-width: 800px;">

- <span style="color: #f686bd">1. Aprendizagem Supervisionada: </span> É a mais utilizada. Funciona como se houvesse um "professor" ensinando o algoritmo: você fornece os dados de entrada já com os respectivos rótulos (respostas). O modelo aprende a relação entre os dados e os rótulos e, no futuro, consegue classificar novos dados sozinho (como uma criança que aprende a identificar um cachorro). **Subdivide-se em**: 

- Classificação: Lida com variáveis categóricas (qualitativas) Exemplo: identificar se uma imagem é de um carro, moto ou lancha, ou classificar se uma notícia é verdadeira ou falsa (Fake News).
- Regressão: Lida com variáveis quantitativas (numéricas) Exemplo: estimar o valor financeiro de uma casa com base em sua localização, número de andares e quantidade de janelas


- <span style="color: #f686bd">2. Aprendizagem Não Supervisionada: </span>  Não possui dados rotulados nem um "professor". É muito utilizada em análise exploratória de dados, onde o algoritmo precisa se virar para encontrar padrões ocultos, agrupando os dados por similaridade ou comportamento (ex: separar imagens de barcos e aviões pelas suas características físicas, sem que ninguém tenha dito o que é cada imagem).

- <span style="color: #f686bd">3. Aprendizado por Reforço: </span> Baseado na interação de um "agente" com um "ambiente". O agente realiza ações e recebe feedbacks (reações) do ambiente, aprendendo por tentativa e erro até descobrir o comportamento ideal. É muito utilizado em robótica e videogames (ex: um robô aprendendo a andar em terreno irregular caindo e levantando, ou um carrinho virtual aprendendo a dar ré para ganhar impulso e subir uma ladeira).

<h2 style="color: #f686bd">Exemplos de Modelos por Tipo de Aprendizado</h2>
 
---


<h4 style="color: #f686bd">Supervisionado — Classificação</h4>
Algoritmos de classificação aprendem a separar dados em categorias discretas. Veja alguns dos mais utilizados:


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center; flex-direction: row-reverse;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">K-Nearest Neighbors (KNN)</strong><br><br>
    Um dos algoritmos mais intuitivos de ML. Para classificar um novo dado, ele olha para os <em>K vizinhos mais próximos</em> no espaço de features e atribui a classe mais votada entre eles. É como perguntar: <em>"quem são meus vizinhos mais parecidos comigo?"</em> e assumir que pertenço ao mesmo grupo. Simples de entender, porém pode ser lento em bases de dados muito grandes.
  </div>
  <img src="imagens/knn.png" width="350" style="border-radius: 10px;">
</div>


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">Árvore de Decisão (Decision Tree)</strong><br><br>
    Funciona como um fluxograma de perguntas: a cada nó, o algoritmo faz uma pergunta sobre uma feature (<em>"a temperatura é maior que 30°C?"</em>) e divide os dados em ramos até chegar a uma folha com a classificação final. É altamente interpretável — você consegue visualizar o raciocínio do modelo — mas pode sofrer overfitting se ficar grande demais.
  </div>
  <img src="imagens/arvores_decicao.png" width="350" style="border-radius: 10px;">
</div>


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center;flex-direction: row-reverse;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">Support Vector Machine (SVM)</strong><br><br>
    O SVM busca o <em>hiperplano</em> que melhor separa classes diferentes, maximizando a margem entre os pontos mais próximos de cada classe (os chamados <em>vetores de suporte</em>). É excelente em problemas de alta dimensionalidade e funciona bem mesmo com poucos dados de treino. Pode ser aplicado em classificação de textos, imagens e dados biomédicos.
  </div>
  <img src="imagens/svm.png" width="350" style="border-radius: 10px;">
</div>

<h4 style="color: #f686bd">Supervisionado — Regressão</h4>
Algoritmos de regressão aprendem a prever valores numéricos contínuos. Veja os principais:


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center; ">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">Regressão Linear</strong><br><br>
    O modelo mais clássico de regressão. Ajusta uma linha reta (ou hiperplano, em múltiplas dimensões) que melhor representa a relação entre as variáveis de entrada e a variável-alvo. É rápido, interpretável e serve como excelente linha de base (<em>baseline</em>) antes de testar modelos mais complexos. Exemplo: prever o preço de uma casa a partir de sua área e número de quartos.
  </div>
  <img src="imagens/regressao_linear.png" width="350" style="border-radius: 10px;">
</div>


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center;flex-direction: row-reverse;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">Random Forest</strong><br><br>
    Ao invés de usar uma única Árvore de Decisão, o Random Forest cria <em>centenas delas</em>, cada uma treinada com uma amostra aleatória dos dados e das features. A previsão final é a média (regressão) ou votação (classificação) de todas as árvores. Essa combinação de muitos modelos fracos gera um modelo robusto com excelente resistência ao overfitting.
  </div>
  <img src="imagens/random_forest.png" width="350" style="border-radius: 10px;">
</div>

<h3 style="color: #f686bd">Não Supervisionado</h3>
Sem rótulos, esses algoritmos encontram estruturas e padrões ocultos nos dados por conta própria:


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">K-Means (Clusterização)</strong><br><br>
    O K-Means divide os dados em <em>K grupos (clusters)</em> com base na similaridade. O algoritmo posiciona K centros aleatórios, associa cada ponto ao centro mais próximo, recalcula os centros e repete até estabilizar. É amplamente usado para segmentação de clientes, agrupamento de documentos e análise de padrões de comportamento.
  </div>
  <img src="imagens/kmeans.png" width="350" style="border-radius: 10px;">
</div>


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0;  justify-content:center;flex-direction: row-reverse;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">PCA — Análise de Componentes Principais</strong><br><br>
    O PCA é uma técnica de <em>redução de dimensionalidade</em>: transforma um conjunto de dados com muitas features em um conjunto menor, preservando ao máximo a variância (a informação) original. É muito usado como pré-processamento para facilitar a visualização e acelerar outros algoritmos. Pense nele como um <em>compressor inteligente</em> de informação.
  </div>
  <img src="imagens/pca.png" width="350" style="border-radius: 10px;">
</div>

<h4 style="color: #f686bd">Aprendizado por Reforço</h4>
O agente aprende por tentativa e erro, acumulando recompensas ao longo do tempo interagindo com o ambiente:


<div style="display: flex; align-items: center; gap: 30px; margin: 20px 0; justify-content:center;">
  <div style="width: 380px; min-width: 380px;">
    <strong style="color: #f686bd; font-size: 1.05em;">Q-Learning</strong><br><br>
    Um dos algoritmos mais fundamentais do aprendizado por reforço. O agente aprende uma tabela de <em>Q-values</em> que associa cada par (estado, ação) a uma recompensa esperada. Com o tempo, ele descobre quais ações levam às maiores recompensas em cada situação. É a base dos algoritmos modernos de IA para jogos — como o AlphaGo e os bots que aprendem a jogar Atari do zero.
  </div>
  <img src="imagens/ql.png" width="350" style="border-radius: 10px;">

<h3 style="color: #f686bd">O Ciclo de Vida de um Projeto de ML e Preparação de Dados</h3>
Na prática, a maior parte do tempo de um projeto (e a etapa mais trabalhosa) é gasta na preparação, limpeza e pré-processamento dos dados, removendo ruídos e retirando partes inúteis.Para treinar um algoritmo, o conjunto de dados precisa ser dividido em três partes (geralmente na proporção de 80% treino e 20% para os demais):

- <span style="color: #f686bd">Dados de Treino: </span> A maior parte dos dados, usada para o algoritmo aprender e identificar padrões.

- <span style="color: #f686bd">Dados de Validação: </span> Usados durante a criação do modelo para atestar se ele está realmente aprendendo (funciona como uma "prova de fim de semestre").

- <span style="color: #f686bd">Dados de Teste: </span> Dados completamente inéditos, usados apenas depois que o modelo final é gerado, para testar a sua performance real e sua capacidade de generalização.

<h3 style="color: #f686bd">Problemas Comuns: Overfitting e Underfitting</h3>

- <span style="color: #f686bd">Underfitting: </span>Ocorre quando o modelo é ruim e não consegue se ajustar nem mesmo aos dados de treino.


- <span style="color: #f686bd">Overfitting: </span> É quando o modelo "decora" excessivamente os dados de treino, mas perde sua capacidade de generalizar. Na prática, ele acerta tudo no treino, mas erra em dados inéditos.


<div style="display: flex; justify-content: center;">
  <img src="imagens/problemas.png" width="1000">
</div>


Para evitar o overfitting, utiliza-se a Regularização, que penaliza os pesos do modelo na função de perda para que não atinjam valores muito altos. Em Redes Neurais, uma técnica popular é o Dropout, que desliga aleatoriamente uma porcentagem de neurônios durante o treinamento, forçando a rede a criar caminhos de aprendizado diferentes.

<h3 style="color: #f686bd">Desafios Práticos da Área</h3>
Ao trabalhar com Machine Learning, os três maiores desafios atuais são:<span style="color: #f686bd"> Quantidade de dados: </span>Redes Neurais precisam de bases de dados gigantescas para funcionar bem. Se há poucos dados disponíveis para treino, é muito melhor utilizar modelos mais simples como SVM ou Random Forest, que desempenham bem em cenários enxutos.<span style="color: #f686bd"> Qualidade dos dados: </span> Lidar com dados não estruturados, informações cheias de ruído e erros ortográficos toma muito tempo e impacta a performance.<span style="color: #f686bd"> Planejamento inadequado: </span> Escolher o algoritmo errado para o problema (ex: usar regressão linear quando um SVM seria mais performático).
